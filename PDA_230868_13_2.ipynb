```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
df_data = pd.read_csv("InterFirmTrade.csv",header = None)
df_data
```

```python
nodes = list(df_data[df_data[0]==13][1]) + list(df_data[df_data[1]==13][0]) + [13]
nodes = list(set(nodes))
df_part = df_data[df_data[0].isin(nodes) & df_data[1].isin(nodes)]
df_part
```

```python
G = nx.from_pandas_edgelist(df_part, source=0, target=1, create_using=nx.DiGraph())
pos = nx.spring_layout(G, k=0.5)
nx.draw(G, pos, with_labels=True)
plt.savefig("PDA_230868_13_2.png")
plt.show()
```
