```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
T=100
alphas=[0,0.5,1]
variances_dict={}

for alpha in alphas:
    samples=np.zeros((1000,T+1))
    for i in range(1000):
        x = [0] * 101
        x[0]=0
        x[1]=0
        for t in range(1,101):
            x[t] = x[t-1] + alpha*(x[t-1]-x[t-2]) + rng.normal(0,1)
        samples[i, :]=x
    variances=np.var(samples, axis=0, ddof=1) #不偏分散
    variances_dict[alpha]=variances

theor=np.arange(T+1) #Var(xt)=t
```

```python
for alpha in alphas:
    plt.scatter(range(T+1), variances_dict[alpha])
plt.plot(range(T+1), theor, linestyle='--', color='black')

plt.xscale('log')
plt.yscale('log')
plt.xlabel("t")
plt.ylabel("Var")
plt.savefig("PDA_230868_11_3.png")
plt.show()
```
