```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
T=100

samples=np.zeros((1000,T+1))
for i in range(1000):
    x = [0] * 101
    for t in range(1,101):
        x[t] = x[t-1] + rng.normal(0,1)
    samples[i, :]=x

variances=np.var(samples, axis=0, ddof=1) #不偏分散
theor=np.arange(T+1) #Var(xt)=t

plt.scatter(range(T+1), variances)
plt.plot(range(T+1), theor, linestyle='--', color='black')
plt.xlabel("t")
plt.ylabel("Var")
plt.savefig("PDA_230868_11_2.png")
plt.show()
```
