```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng()
```

```python
u=rng.uniform(0, 1, 10000)
x=1/np.sqrt(1-u)

bins = np.logspace(0, 2, 21)
bin_centers = 0.5*(bins[1:] + bins[:-1])
hist,_ = np.histogram(x, bins, density = True)
plt.scatter(bin_centers, hist)

x_val = np.logspace(0, 2, 301)
y_val=2*np.power(x_val,-3)
plt.plot(x_val,y_val,linestyle='--',color='black')
plt.xscale('log')
plt.yscale('log')
plt.title("PDF of Beki Distribution")
plt.savefig("PDA_230868_10_2.png")
plt.show
```
