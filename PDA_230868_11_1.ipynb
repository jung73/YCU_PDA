```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
for i in range(10):
    x = [0] * 101
    for t in range(1,101):
        x[t] = x[t-1] + rng.normal(0,1)
    plt.plot(x)

plt.title("Sample Path (10 Path)")
plt.xlabel("t")
plt.ylabel("x")
plt.savefig("PDA_230868_11_1.png")
plt.show()
```
